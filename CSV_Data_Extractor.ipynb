```python
# Import numpy, pandas, and os libraries 
import numpy as np 
import pandas as pd 
import os

# load CSV files 
csv_directory = os.getcwd()
csv_files = [f for f in os.listdir(csv_directory) if f.endswith('.csv')]

#Create a new folder called Output in the current working directory if it doesn't exist 
output_file_path = os.path.join(csv_directory, 'Output')
if not os.path.exists(output_file_path):
    os.makedirs(output_file_path)

for csv in csv_files:
    data = pd.read_csv(csv)  #Read CSV files 
    line_id = data.iloc[0,2] #Select line id 
    line_id = line_id[11:] #Strip unwanted data from line ID 
    data['ID'] = line_id # Insert a new column with line ID 
    data1 = data.drop(index = 0) # Delete first row with unwanted data and create a new dataframe 
     
    # Export the Columns of interest to CSV
    MPES = data1[['Position: MPES: East', 'North.6', 'ID']]
    MPES_file_name = os.path.join(output_file_path, line_id + '_MPES.csv')
    MPES.to_csv(MPES_file_name, index=None)

    SBES = data1[['Position: SBES: East', 'North.7', 'ID']]
    SBES_file_name = os.path.join(output_file_path, line_id + '_SBES.csv')
    SBES.to_csv(SBES_file_name, index=None)
```
